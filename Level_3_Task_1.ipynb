```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
df.head(1)
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df = df[df['Aggregate rating'] !=0]
```

```python
df.shape
```

Task 1

Predictive Modeling


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Has Table booking"] = le.fit_transform(df["Has Table booking"])
df["Has Online delivery"] = le.fit_transform(df["Has Online delivery"])
df["Is delivering now"] = le.fit_transform(df["Is delivering now"])
df["Switch to order menu"] = le.fit_transform(df["Switch to order menu"])
```

```python
df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["City"] = le.fit_transform(df["City"])
df["Cuisines"] = le.fit_transform(df["Cuisines"])
df["Rating color"] = le.fit_transform(df["Rating color"])
df["Rating text"] = le.fit_transform(df["Rating text"])
df["Currency"] = le.fit_transform(df["Currency"])
df["Restaurant ID"] = le.fit_transform(df["Restaurant ID"])
df["Country Code"] = le.fit_transform(df["Country Code"])
df["Restaurant Name"] = le.fit_transform(df["Restaurant Name"])
df["Address"] = le.fit_transform(df["Address"])
df["Locality"] = le.fit_transform(df["Locality"])
df["Locality Verbose"] = le.fit_transform(df["Locality Verbose"])
```

```python
x = df.drop(columns = ["Aggregate rating"])
y = df["Aggregate rating"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
r2_score(y_test, model.predict(x_test))
```

```python
mean_squared_error(y_test, model.predict(x_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
model2 = DecisionTreeRegressor()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
r2_score(y_test, model2.predict(x_test))
```

```python
mean_squared_error(y_test, model2.predict(x_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
model3 = RandomForestRegressor()
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
r2_score(y_test, model3.predict(x_test))
```

```python
mean_squared_error(y_test, model3.predict(x_test))
```
